# Importing of the needed library 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# The datasets of both Days and Temperature in a list format

```python
Days = [1,2,3,4,5,6,7]
Temperature =[50,51,52,48,47,49,46]
```

# making of different visualization of days and temperature on x and y axis

```python
plt.plot(Days,Temperature, color="green")
plt.xlabel("Days", fontsize=15)
plt.ylabel("Temperature", fontsize=15)
plt.title("Weather chart")
plt.grid()
```

```python
plt.plot(Days,Temperature, color="green", linewidth=2, linestyle="dashed")
plt.xlabel("Days", fontsize=15)
plt.ylabel("Temperature", fontsize=15)
plt.title("Weather chart")
plt.grid()
```

```python
plt.plot(Days,Temperature, color="red", linewidth=2, linestyle="dotted")
plt.xlabel("Days", fontsize=15)
plt.ylabel("Temperature", fontsize=15)
plt.title("Weather chart")
plt.grid()
```

```python
plt.plot(Days,Temperature, color="blue", linewidth=2, linestyle="-.")
plt.xlabel("Days", fontsize=15)
plt.ylabel("Temperature", fontsize=15)
plt.title("Weather chart")
plt.grid()
```

```python
plt.bar(Days,Temperature, color="purple")
plt.xlabel("Days", fontsize=15)
plt.ylabel("Temperature", fontsize=15)
plt.title("Weather chart")
plt.grid()
```

```python
plt.boxplot(Days,Temperature)
plt.xlabel("Days", fontsize=15)
plt.ylabel("Temperature", fontsize=15)
plt.title("Weather chart")
plt.grid()
```

```python
plt.pie(Days,Temperature)
plt.xlabel("Days", fontsize=15)
plt.ylabel("Temperature", fontsize=15)
plt.title("Weather chart")
```

## From the above chart, we can observe the following through all the chart drawn above
### (1) The maximum temperature on the third day
### (2) The minimum temperature on the seventh day
