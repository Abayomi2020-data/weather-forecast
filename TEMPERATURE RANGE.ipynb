## Importing of needed library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## The dataset of Days, minimun temperature, maximum temperature, average temperature used in the line graph

```python
Days = [1,2,3,4,5,6,7]
Max_t = [50,51,52,48,47,49,46]
Min_t = [43,42,40,44,33,35,37]
Avg_t = [45,48,48,46,40,42,41]
```

```python
plt.title("Weather", fontsize=14)
plt.xlabel("Days", fontsize=14)
plt.ylabel("Temperature", fontsize=14)
plt.plot(Days,Max_t, label="Max")
plt.plot(Days,Min_t, label="Min")
plt.plot(Days,Avg_t, label="Avg")
```

## The line graph above shows different ranges of temperature, which are minimum, maximum and average temperature without specifying the line that shows each of the lines which are minimum, maximum and average 

```python
plt.title("Weather", fontsize=14)
plt.xlabel("Days", fontsize=14)
plt.ylabel("Temperature", fontsize=14)
plt.plot(Days,Max_t, label="Max")
plt.plot(Days,Min_t, label="Min")
plt.plot(Days,Avg_t, label="Avg")
plt.legend(loc="best", shadow=True)
```

## Legend was added to the line graph above to specify what each line stand for with correct label 

```python
plt.title("Weather", fontsize=14)
plt.xlabel("Days", fontsize=14)
plt.ylabel("Temperature", fontsize=14)
plt.plot(Days,Max_t, label="Max")
plt.plot(Days,Min_t, label="Min")
plt.plot(Days,Avg_t, label="Avg")
plt.legend(loc="best", shadow=True)
plt.grid()
```
